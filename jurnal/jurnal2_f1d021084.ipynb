a. translasi sejauh x = -50 dan y = 40

Membaca citra 'beanny.jpg' dalam mode grayscale dan menyimpannya dalam variabel 'draw' menggunakan fungsi cv.imread('beanny.jpg',0)

Menghitung panjang (jumlah baris) dari citra 'draw' menggunakan fungsi len(draw)

Mengubah citra 'draw' menjadi array NumPy menggunakan fungsi np.array(draw)

Membuat matriks nol dengan ukuran (height, width) yang sama dengan citra 'draw' menggunakan fungsi np.zeros((height,width)).

Melakukan translasi pada citra 'draw' sebesar 50 piksel ke kanan dan 40 piksel ke atas pada setiap iterasi looping, nilai variabel 'i' dan 'j' akan digunakan sebagai indeks untuk mengakses setiap piksel pada citra 'draw'. Selanjutnya, nilai 'i' dan 'j' akan digunakan untuk menghitung koordinat piksel baru setelah dilakukan translasi dengan menggunakan rumus
nilai 'i' dan 'j' ditambah atau dikurangi dengan nilai translasi yang telah ditentukan. Kemudian, hasilnya dibatasi dengan ukuran citra menggunakan operasi modulo (%) dengan nilai 'height' dan 'width'. Hal ini dilakukan agar koordinat piksel yang dihasilkan tetap berada dalam batas citra yang ada.

Setelah itu, akan dilakukan pengecekan apakah koordinat piksel baru berada dalam batas citra yang diinginkan




```python
draw = cv.imread('beanny.jpg',0)
leng = len(draw)
draw = np.array(draw)
translation = np.zeros((height,width))

for i in range(leng-50):
    for j in range(leng-50):
            x = (i + 50) % height
            y = (j - 40) % width
            if x < height + 50 and y < width - 40:
                 translation[x][y] = draw[i][j]
            
mp.subplot(1, 2, 1)
mp.title('Before Transtation')
mp.imshow(grayImg, cmap='gray')

mp.subplot(1, 2, 2)
mp.title('After Transtation')
mp.imshow(translation, cmap='gray')

print(draw)
```

b. rotasi dengan sudut 70 derajat

=== Rumus Transformasi Geometri ====
 
Rotasi dengan Titik Pusat (0,0) dengan Sudut Putar α:

x’ = x cos α – y sin

y’ = x sin α + y cos α

Mendeklarasikan nilai matriks dalam foto
.Membuat sebuah variabel kosong untuk menampung hasil rotasi
.Dibuat variabel putaran dengan rumus np.pi sebagai konstanta dari library NumPy ang merepresentasikan nilai pi dalam radian, dan 180 digunakan untuk mengkonversi derajat menjadi radian dan akhirnya akan dikalikan dengan 70 untuk mendapatkan hasil yang berputas searah jarum jam.

x//2 digunakan untuk mendapatkan nilai tengah dari dimensi citra tersebut, yang akan digunakan sebagai titik pusat rotasi.
fungsi // adalah untuk membagi nilai dan dibulatkan kebawah

Dibentuk fungsi sum(round) untuk setiap masukkan nilai rotasi positif atau negatif akan mempengaruhi rumus dalam fungsi

Melakukan perulangan untuk mengecek pixel demi pixel secara horizontal dan vertikal dengan variabel untuk mengecek yaitu x dan y

i dan j adalah koordinat x dan y dari titik yang akan dirotasi.
NoteX dan noteY adalah koordinat x dan y dari titik pusat rotasi.
round adalah sudut rotasi dalam radian yang akan diterapkan pada titik-titik tersebut.
np.cos dan np.sin adalah fungsi kosinus dan sinus dalam modul NumPy.

x2 dan y2 adalah koordinat x dan y dari titik yang sudah dirotasi.
x dan y adalah ukuran panjang dan lebar bidang 2D.
round adalah sudut rotasi dalam radian yang akan diterapkan pada titik-titik tersebut.
rotation1 dan rotation2 adalah array dua dimensi yang merepresentasikan gambar yang sudah dirotasi.






```python
x, y = draw.shape

rotation1 = np.zeros((x, y))
rotation2 = np.zeros((x, y))

roundClock = 70*(np.pi/180)
roundNonClock = -70*(np.pi/180)

noteX = x // 2 
noteY = y // 2

def sum(round):
    for i in range(x):
        for j in range(y):
            x2 = int((i - noteX) * np.cos(round) - (j - noteY) * np.sin(round) + noteX)
            y2 = int((i - noteX) * np.sin(round) + (j - noteY) * np.cos(round) + noteX)
            if 0 <= x2 < x and 0 <= y2 < y:
                if round > 0: 
                    rotation1[i][j] = draw[x2][y2]
                else:
                    rotation2[i][j] = draw[x2][y2]
                    
mp.subplot(1, 2, 1)
mp.title('Positif Rotation')
sum(roundClock)
mp.imshow(rotation1, cmap="gray")

mp.subplot(1, 2, 2)
mp.title('Negatif Rotation')
sum(roundNonClock)
mp.imshow(rotation2, cmap="gray")
```

Pada setiap iterasi loop kedua, loop kedua "j" digunakan untuk menyalin elemen gambar pada baris "i" dari "draw" ke baris "i" dari "ZeroMatr2". Perhatikan bahwa nilai "mirror" digunakan untuk mengakses elemen gambar pada kolom "mirror" dari "draw", sehingga membuat gambar di "ZeroMatr2" terlihat seperti dicerminkan secara vertikal.

```python
ZeroMatr = np.zeros((height,width))
ZeroMatr2 = np.zeros((height,width))
mirror = leng-1

for i in range(height):
        for j in range(width):
                ZeroMatr[i][j]=draw[mirror][j]
        mirror-=1

for i in range(height):
        mirror = leng-1
        for j in range(width):
                ZeroMatr2[i][j]=draw[i][mirror]
                mirror-=1

mp.subplot(1, 2, 1)
mp.title('X')
mp.xticks([]), mp.yticks([])
mp.imshow(ZeroMatr, cmap='gray')

mp.subplot(1, 2, 2)
mp.title('Y')
mp.xticks([]), mp.yticks([])
mp.imshow(ZeroMatr2, cmap='gray')

print(draw)
```

Loop pertama digunakan untuk mengakses setiap baris pada matriks gambar "draw" menggunakan variabel "i". Loop kedua digunakan untuk mengakses setiap kolom pada matriks gambar "draw" menggunakan variabel "j". Loop ketiga dan keempat digunakan untuk mengakses setiap pixel pada setiap grid 3x3 menggunakan variabel "k" dan "l".

nilai dari elemen gambar pada baris "i" dan kolom "j" pada matriks gambar "draw" akan disalin ke elemen gambar pada baris "skalai+k" dan kolom "skalaj+l" pada matriks gambar "ZeroMatr3". Hal ini dilakukan untuk memperbesar gambar sebanyak tiga kali lipat.

Setelah selesai melakukan operasi dilatasi, gambar pada matriks gambar "ZeroMatr3" ditampilkan pada plot menggunakan method "imshow" dari "mp" dengan cmap 'gray', dan gambar pada "draw" juga dicetak di console menggunakan fungsi "print".

```python
skala = 3
dilation = leng*3
ZeroMatr3 =np.zeros((dilation,dilation))

for i in range(height):
    for j in range(width):
        for k in range(skala):
            for l in range(skala): 
                ZeroMatr3[skala*i+k,skala*j+l]=draw[i,j]
mp.imshow(ZeroMatr3, cmap="gray")
print(draw)
```

2. Lakukan Operasi Citra Negatif dari citra grayscale lalu bandingkan Histogram kedua citra tersebut.

Untuk mengubah gambar grayscale menjadi negatif adalah mengurangi indeks gambar dengan 255.
Nilai 255 pada imgNeg merepresentasikan nilai maksimum intensitas piksel pada citra grayscale (dalam skala 8-bit) yang dijadikan acuan untuk proses inversi. Dalam proses inversi, nilai intensitas piksel pada citra grayscale akan dikurangkan dari nilai maksimum, sehingga hasilnya akan menjadi kebalikan dari nilai awal.

Melakukan perulangan untuk mengecek pixel demi pixel secara horizontal dan vertikal dengan variabel untuk mengecek yaitu tinggi dan lebar kemudian Setiap pixel pada histGray akan diisi oleh tiap pixel grayImg

Setiap pixel pada histNeg akan diisi oleh tiap pixel grayNeg

Tampilkan hasil untuk melihat perbandingan
subplot adalah library numpy untuk menentukan lokasi output gambar dengan koordinat yang sudah ditentukan
Diberikan Judul Image
Dan xticks dan yticks adalah fungsi yang menghilangkan angka koordinat pada tabel histogram




```python
imgNeg = (255 - grayImg)

histGray = np.zeros((height,width))

for i in range(height):
    for j in range(width):
        histGray[i,j] = grayImg[i,j]

histNeg = np.zeros((height,width))
for i in range(height):
    for j in range(width):
        histNeg[i,j] = imgNeg[i,j]

mp.subplot(1, 2, 1)
mp.title('GrayScale')
mp.xticks([]), mp.yticks([])
mp.imshow(grayImg, cmap='gray')

mp.subplot(1, 2, 2)
mp.title('NegatifScale')
mp.xticks([]), mp.yticks([])
mp.imshow(imgNeg, cmap='gray')
```

=== HASIL PERBANDINGAN BAR ===

v.calcHist() adalah fungsi yang digunakan untuk menghitung histogram dari citra.
indeks channel pada citra (0 untuk grayscale, 0, 1, atau 2 untuk RGB).
None disitu citra mask (opsional) yang digunakan untuk menghitung histogram hanya pada area tertentu dari citra.
256 jumlah bin (interval nilai intensitas) yang akan digunakan pada histogram.
[0,256] adalah rentang nilai intensitas piksel pada citra.

mp.bar() adalah fungsi yang digunakan untuk membuat histogram
hist_neg.flatten() digunakan untuk mengubah hasil perhitungan histogram dari masing-masing citra menjadi array 1 dimensi, sehingga bisa diproses oleh fungsi mp.bar().

mp.bar() adalah fungsi yang digunakan untuk membuat histogram



```python
hist_gray = cv.calcHist([imgNeg], [0], None, [256], [0, 256])
hist_neg = cv.calcHist([grayImg], [0], None, [256], [0, 256])

mp.bar(np.arange(256), hist_gray.flatten(), color='grey')
mp.bar(np.arange(256), hist_neg.flatten(), color='orange', alpha=0.7)

mp.legend(('Citra Asli', 'Citra Negatif'), loc='upper left')
mp.show()
```
